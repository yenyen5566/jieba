```python
def make_txt(abc):
    for i in abc:
        with sqlite3.connect('mining.sqlite') as conn: 
            c = conn.cursor()
            result = list(c.execute('select * from {} ;'.format(i)))

        with open('{}.csv'.format(i),'w') as f:
            for r in result:
                final = str(r[1]) + ',' + str(r[2]) + '\n'
                f.write(final)
                
```

```python
def ptt_content(url,table_name): #解析出內文並回存進資料庫
    import sqlite3
    import requests
    from bs4 import BeautifulSoup
    import time
    from datetime import date
    
    try:

        time.sleep(0.3)
        head = {"cookie": "over18=1;"}
        
        res = requests.get(url ,headers = head)
        
        soup = BeautifulSoup(res.text,'lxml')
        
        catched = []

        title = soup.select('div.article-metaline > span')[3].text #title

        catched.append(title)

        pushs = soup.select('.push-content')

        for push_content in pushs:
            text = push_content.text.replace(':','')
            catched.append(text)

        for trash in soup.select('div.article-metaline > span') + soup.select('div.article-metaline-right > span') + soup.select('div.push'):
            trash.decompose()

        content_raw = soup.select_one('#main-content').text.strip()
        content = content_raw.split('\n※ 發信站')[0].replace('\n','')

        catched.append(content)    
        
        with sqlite3.connect('mining.sqlite') as conn:
            c = conn.cursor() 
            save_soup = "update {} set content = ? ,getTime = ? where url = ?".format(table_name) 
            c.execute(save_soup,(content,str(date.today()),url))
    except IndexError as e:
        print(e, url)
    
class CrawlerError(Exception): 
    pass
```
