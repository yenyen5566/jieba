```python
def wordCount(filename):  
    ''''' 
    简单计算词频的函数 
    :param filename: 文件名 
    :return: 词频 
    '''  
    wordCount = {}  
    file = open(filename,'r')  
    while True:  
        line = file.readline()  
        if line:  
            wordlist = line.split(',[')  
            if len(wordlist) == 1: continue  
            wordlist = wordlist[1].split('],')[0].split(',')  
            for word in wordlist:  
                word = word.replace(' ', '').replace("'", '')  
                if word in wordCount.keys():  
                    wordCount[word] = wordCount[word]+1  
                else:  
                    wordCount[word] = 1  
        else:  
            break  
    return [(k, wordCount[k]) for k in wordCount.keys()]  
```

```python
def word_count(location):
    
     
    import jieba
    import re
    import jieba.analyse  #載入能萃取出文章中關鍵詞的功能
    jieba.set_dictionary('dict.txt')

    stop = []                                    
    stops = open('stop2.txt','r',encoding='utf8').readlines()
    for s in stops:
        stop.append(s.strip())

    for x in location:
        ret = open("./txt/{}.txt".format(x), "r",encoding='utf8').read()
        words = seglist = jieba.cut(ret, cut_all=False)

        content = []

        for word in words:
            if word not in stop:
                content.append(word)

        hash = {}
        for item in content: 
                if item in hash:
                      hash[item] += 1
                else:
                      hash[item] = 1
        with open('./word_count/{}.csv'.format(x),'w',encoding='utf8') as f:
                for i in hash:
                    f.write( i +','+str(hash[i])+'\n')
```
