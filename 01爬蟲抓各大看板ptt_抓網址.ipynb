```python
for i in all_data:    
    with sqlite3.connect('mining.sqlite') as conn:
        c = conn.cursor()
        try:
            c.execute('drop table {}'.format(i)) #換
            print('drop table {}'.format(i)) #換
        except:
            print(1)
        c.execute('''
            CREATE TABLE {}(   
              url text,
              title text,
              content text,
              reply text,
              getTime datetime)'''.format(i))
        print('create table {}'.format(i)) #換
for i in alll:    
    wrong = ['Wen-Shan','San-Ying','Shu-Lin','WuGu-BaLi']
    right = ['Wen_Shan','San_Ying','Shu_Lin','WuGu_BaLi']
    if i in wrong:
        for r in right:
            z = i.split('-')[0]
            k = r.split('_')[0]
            if z == k:    
                place = i 
                place2 = r
                insert_sqlite(place,place2)
    else:
        place = i 
        place2 = i
        insert_sqlite(place,place2)
```

```python
def insert_sqlite(place,place2):
    
    
    HOST = "https://www.ptt.cc"
    BOARD = "/bbs/{}/index.html".format(place)  #換
    INDEX = "/bbs/{}/".format(place)
    INDEX1 = "index{}.html" #換
    
    total_url = 0
    
    try:
        
        HOST = "https://www.ptt.cc"
        headers = {"cookie": "over18=1;"}
        res = requests.get(HOST + BOARD, headers=headers)
        soup = BeautifulSoup(res.text, 'lxml')
        buttons = soup.select('a.btn.wide')
        total_page = int(buttons[1]['href'].split('index')[1].split('.html')[0]) + 1
        
        
        with sqlite3.connect('mining.sqlite') as conn:  #連線資料庫 
            c = conn.cursor()
            
            page_to_crawl = 30
            for page in range(total_page, total_page - page_to_crawl, -1):
                res = requests.get(HOST +INDEX+ INDEX1.format(page), headers=headers)
                soup = BeautifulSoup(res.text, 'lxml')
                links = soup.select('div.title > a')
                for link in links:
                    url = HOST + link['href']
                    title = link.text
                    insert_string = "INSERT INTO {}(title, url) VALUES (?, ?)".format(place2) #換
                    c.execute(insert_string, (title, url))
                    print(url)
                    time.sleep(0)

                    total_url += 1
                
    except Exception as ex:
        print(ex)
        

    print('{} url was catched'.format(total_url))
    
```
